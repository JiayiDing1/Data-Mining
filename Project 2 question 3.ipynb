```python
def introduction():
    print('''The main function will ask the user to type three integer values, and will print all three numbers.
    Then it will call the funtion findaverage to determine the average of these vales.
    The main function will print the value returned, with three decimal places.
    Then the main function will call comparetoave function to compare each of the three parameter values to the average. 
    The function will print an appropriate message for each one: equal to, less than, or greater than the average.
    The function will also print a number of how many of the three original numbers are equal to the average.
    Then the main function will skip a few lines and ask the user to input next set of three numbers.
    In the end, total number of data sets processed are entered.
    ''')
```

```python
# This function takes three numbers from main function(list x) and determine the average of these vales.
def findaverage(x):
    # varialbe ave will be used in other functions
    global ave
    # calculating ave and return result to the main function
    total = x[0]+x[1]+x[2]
    ave = total/3
    return ave
```

```python
# This function takes three numbers from main function(list x) and compare each of the three parameter values to the average.
def comparetoavg(x,ave):
    # vairable equal is used to accumulate number of data that are equal to the average
    equal = 0
    for i in x:
        #The function will print an appropriate message for each number of list x (less than, greater than, and equal to)
        if i > ave:
            print(i,'is greater than average')
        elif i < ave:
            print(i,'is less than average')
        else:
            # else is when i = ave
            print(i,'is equal to average')
            equal +=1
    # print a number of how many of the three original numbers are equal to the average.
    print('There are',equal,'numbers equal to the average')
```

```python
def main():
    introduction()
    # Make sure there are at least 10 datasets
    for i in range(0,11):
        # append list x with three input numbers
        x=[]
        for n in range(3):
            # input 3 numbers
            n = int(input('Integer:'))
            # add all 3 into list x
            x.append(n)
        print('The three original integers are:',x[0],x[1],x[2])
        findaverage(x)
        #The main function will print the value returned, with three decimal places.
        # formation of average learned from stackflow
        print('The average is:','{:.3f}'.format(ave))
        
        comparetoavg(x,ave)
        print('\t')
        print('\t')
        # ask the user to input next set of three numbers
        i = i+1
    print(i,'sets of data is processed')
```
