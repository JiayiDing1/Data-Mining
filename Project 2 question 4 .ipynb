```python
def introduction():
    print('''This function will read in three integers representing a student's scores. 
    It will filter them by if they are valid, if they are the same, and finally rounding them. 
    If it is a valid set of scores, it will print that it is valid set of score, and print invalid if it is not.
    It will then compare the score to each other into four senarios: 
    all different, all the same, 2 are same and the third is larger, 2 are same and the third is smaller.
    Tt will then proceed and rounding original score into 4 decimal places.
    It will then compare the score to each other for the second time.
    If it is a invalid set of scores, it will read the next set.
    It will process all 20 datasets''')
```

```python
#The function will read list x from main function and determine whether or not the three scores are valid: in the range from 0 - 100

def aretheyvalid(i, x):
    # variable c is used to determing if a set of number is valid or invalid. If it is invalid it will add 1 to current value.
    c = 0
    for num in range(3*i-3,3*i):
        # if the score is less than 0, it is invalid because it is too small
        if x[num] < 0:
            # accumulating number of invalid scores
            c += 1
            print('Number',x[num], 'invalid because it is too small')
            print('This a invalid set of scores')
            print('\t')
            # boolean vairable returned to main function
            return False
        # if the score is greater than 100, it is invalid because it is too big
        elif x[num] > 100:
            # accumulating number of invalid scores
            c += 1
            print('Number',x[num], 'invalid because it is too big')
            print('This a invalid set of scores')
            print('\t')
            # boolean vairable returned to main function
            return False
    # It is valid if it does not meet invalid conditions
    if c == 0:
        print('This is a valid set of score')
        # boolean vairable returned to main function
        return True
```

```python
# This function will read list x and y from main function and compare three numbers to determine if it is the same
# There are going to be four senarios returned to the main function
def aretheysame (j, x):
    # i is the index
    i = 3*j - 3
    if x[i] == x[i+1] == x[i+2]:
        print('All three numbers are the same')
    elif x[i] == x[i+1] < x[i+2] or x[i+1] == x[i+2] < x[i] or x[i] == x[i+2] < x[i+1]:
        # three conditions
        print("Two numbers are the same and the third is larger")
    elif x[i] == x[i+1] > x[i+2] or x[i+1] == x[i+2] > x[i] or x[i] == x[i+2] > x[i+1]:
        # three conditions
        print('Two numbers are the same and the third is smaller')
    elif x[i] != x[i+1] != x[i+2]:
        print('All three numbers are different.')
```

```python
# This function read list x from main function and round the integer to the nearest 10; round up for 5 to 9, and round down for 0 to 4.
# It will return to main function the rounded number
def roundscore (i,x, y):
    for num in range(3*i-3,3*i):
        v = x[num]
        # variable m is the decimal place
        m = v % 10
        if m > 4:
            # round up if the number is 5-9
            rounded = v-m+10
            print('Original Number:',v, 'Rounded Number :', rounded)
        else:
            # round down if the number is 0-4
            rounded = v-m
            print('Original Number:',v, 'Rounded Number :', rounded)
        
        # make rounded number into a new list
        y[num] = rounded
```

```python
# main function that connnect all functions
def main ():
    introduction()
    print('\t')
    s = 1
    # valid/invalid is used to accumulate
    valid = 0
    invalid = 0
    # two list: one original and one rounded
    x = []
    y = []
    # read .txt file and make three numbers into a set
    file1 = open("question_4.txt", 'r')
    i = file1.readline().rstrip()
    while i != '':
        j = i.split(" ")
        # if i is not bland (reads the data)
        for num in range (3):
            a = int(j[num])
            print('The test score is:', a)
            x.append(a)
            y.append(a)    
        
        if aretheyvalid(s,x):
            valid += 1
            print('\t')
            aretheysame(s,x)
            print('\t')
            roundscore(s,x,y)
            print('\t')
            # compare again after rounding
            aretheysame(s,y)
            print("\t")
            print('\t')
            print('\t')
            i = file1.readline().rstrip()
            s += 1
        # read the next set of data of it is invalid
        else:
            invalid += 1
            print('\t')
            print('\t')
            i = file1.readline().rstrip()
            s += 1
            
    file1.close()

    print('Total sets: ',s)
    print('Valid: ', valid)
    print('invalid: ', invalid)

```
