```python
import numpy as np
import pandas as pd
```

```python
# read .csv file into dataframe
df = pd.read_csv('question2_data.csv',header = None)
df
```

```python
# The data starts from 0 to 11. It automatically ends after reaching 12.
for i in range(0,12):
    print('Team',int(df.iloc[[i],[0]].values),'has',int(df.iloc[[i],[1]].values),'wins,',
          int(df.iloc[[i],[2]].values),'loses, and',int(df.iloc[[i],[3]].values),'ties')
    total_played = int(df.iloc[[i],[1]].values + int(df.iloc[[i],[2]].values) + int(df.iloc[[i],[3]].values))
    print('Total number of games played:',total_played)
    
    # Season is finished if the team played 16 games already
    if total_played == 16:
        print('Season finished')
    else:
        print('Number of games remaining:',16-total_played)
        
    # Calculating winning average
    if int(df.iloc[[i],[1]].values) != 0:
        print('Winning Average is','{:.4f}'.format(int(df.iloc[[i],[1]].values)/total_played))
    else:
        print('Winning Average is','0.0000')
    
    # Comparing if number of games tied is greater than or equal to the number won
    if int(df.iloc[[i],[3]].values) >=int(df.iloc[[i],[1]].values):
        print('Number of games tied is greater than or equal to the number won')
    else:
        print('Number of games tied is not greater than or equal to the number won')
    
    # Comparing if number of games tied is greater than the number lost
    if int(df.iloc[[i],[3]].values) > int(df.iloc[[i],[2]].values):
        print('Number of games tied is greater than the number lost')
    else:
        print('Number of games tied is not greater than the number lost')
        
    # Calculating wip total for each team
    wip_total = int(df.iloc[[i],[1]].values) + int(df.iloc[[i],[3]].values) - 3*int(df.iloc[[i],[2]].values)
    if wip_total < 0:
        print('Wip Total is', 0)
    else:
        print('wip Total is',wip_total)
        
    print('\t')
# printing total number of teams at the end
print('The total number of teams:',i+1)
```
